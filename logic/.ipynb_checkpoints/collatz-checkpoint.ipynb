```python
#!/opt/anaconda3/bin/python
"""collatz.py
takes in a positive integer `n` as an argument and applies the collatz rules:

1. if `n` is even, half it
2. else set `n=k` where `n=3k+1`

"""
def is_even(n):
  return n % 2 == 0


def collatz(n, meta=False):
  current = n
  path = [n]
  while current != 1:
    if is_even(current):
      current = current // 2
    else:
      current = (current * 3) + 1
    path += [current]

  if meta:
      return path, len(path), max(path)
  return path
```

```python
  t = collatz(7, True)
  print(f"Sequence:\n{t[0]}\n\nLength: {t[1]}\n\nState Space Highest Reached: {t[2]}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
yy = list(map(collatz, np.arange(1,31)))
np.arange(1,31)
```

```python
fig, axs = plt.subplots(2,3, figsize=(10,10))

def line_plot(n, ax):
    for i in range(n):
        xx = np.arange(len(yy[i]))
        ax.plot(xx, yy[i])
        ax.set_title(f"To {n}")
        
for i in range(2):
    for j in range(3):
        line_plot(5 + 15 * i + 5 * j, axs[i, j])

plt.tight_layout()
plt.show()
```
